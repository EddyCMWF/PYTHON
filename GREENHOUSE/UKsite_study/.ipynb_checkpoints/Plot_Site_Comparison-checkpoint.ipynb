```python
#import os
import netCDF4 as nc
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt


#
#import iris
#
#from iris_jules import jules as irju
#from iris_jules import convertpoints2grid as conpts2grid
#
#
from matplotlib.pyplot import rcParams
%matplotlib inline
rcParams['figure.figsize'] = (15.,10)

```

```python
# Directories and Filenames

site='Brattleby'
JULESsubdir='Jvn4.6_variableRGROW'
runids=['r0%02d'%rgrow for rgrow in range(5,31,5)]
tstep='day'
nJULES=len(runids)

BASE_DIR='/prj/GREENHOUSE/GREENHOUSE_sites/'

JULES_OUTPUT_DIR=BASE_DIR+'output/'+site+'/'+JULESsubdir+'/'
JULES_fnames=[JULES_OUTPUT_DIR+site+'_'+runid+'.'+tstep+'.nc' for runid in runids]

SITE_DIR=BASE_DIR+'data/'+site+'/'
SITE_fname=SITE_DIR+'ConCrop_FullFlux_data.nc'

```

```python

WK_varnames=['NEE','GPP','TER','SHF','LHF','T_soil','SMC','SHF_grnd']

SITE_varnames=['NEE','GPP','TER','H','LE','T_soil','SMC','downward_heat_flux_at_ground_level']
SITE_timename='time'

JULES_varnames=['npp_gb','gpp_gb','resp_s_gb','resp_p_gb','t_soil','smcl','fqw_gb','latent_heat']
JULES_timename='time'

```

```python
# Read in Site data and convert to pandas
Sinf=nc.Dataset(SITE_fname,'r')

SITE_Dict={}
for inname,wkname in zip(SITE_varnames,WK_varnames):
    #print(inname,wkname)
    SITE_Dict[wkname]=Sinf.variables[inname][:].squeeze()

SITE_time=nc.num2date(Sinf.variables[SITE_timename][:],  \
                      Sinf.variables[SITE_timename].units)
Sinf.close()

SITE_Dict['T_soil']  = np.mean(SITE_Dict['T_soil'].reshape(-1,4),axis=1)
SITE_Dict['SMC']     = np.mean(SITE_Dict['SMC'].reshape(-1,4),axis=1)
SITE_Dict['SHF_grnd']= np.mean(SITE_Dict['SHF_grnd'],axis=1)

SITE_panda = pd.DataFrame(SITE_Dict,index=SITE_time)
```

```python
# Read in JULES data and convert to pandas
J_pandas=[]

for Jcnt in range(nJULES):
    #print(runids[Jcnt])
    Jinf=nc.Dataset(JULES_fnames[Jcnt],'r')
    
    J_Dict={}
    for inname in JULES_varnames:
        #print(inname)
        J_Dict[inname]=Jinf.variables[inname][:].squeeze()
    
    J_time=nc.num2date(Jinf.variables[SITE_timename][:],  \
                       Jinf.variables[SITE_timename].units)
    Jinf.close()
    
    # Copy some variables straight over
    invars=['npp_gb','gpp_gb','resp_s_gb','resp_p_gb','fqw_gb','latent_heat']
    wkvars=['NPP','GPP','RESP_S','RESP_P','SHF','LHF']
    JULES_Dict={wkvar:J_Dict[invar] for wkvar,invar in zip(wkvars,invars)}
    
    # NEE=NPP-RESP_S
    JULES_Dict['NEE']=JULES_Dict['NPP']-JULES_Dict['RESP_S']
    # TER = RESP_S + RESP_P
    JULES_Dict['TER']=JULES_Dict['RESP_S']+JULES_Dict['RESP_P']
    
    # Top soil layer (top 10cm)
    JULES_Dict['T_soil'] = J_Dict['t_soil'][:,0]
    JULES_Dict['SMC'] = J_Dict['smcl'][:,0]
    J_pandas.append( pd.DataFrame(JULES_Dict,index=J_time) )
```
