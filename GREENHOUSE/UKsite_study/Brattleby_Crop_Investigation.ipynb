```python
########################################
# Constants and options
kgC_to_umolsCO2_factor =  (1e3/12.0107)*1e6  #(grams of substance/molar mass) * micromoles conversion
seconds_in_year = 60.*60.*24.*365.
Lc_H2O = 2.501e6
```

```python
# Directories and Filenames

PLOT_DIR='/prj/GREENHOUSE/GREENHOUSE_sites/plots/Brattleby/crop_tuning/'

site='Brattleby'
JULESvn = 'Jvn4.6'
runids=['crop_albmar','crop_albmar_mod1','crop_albmar_mod2','crop_albmar_mod3', \
        'crop','crop_mod1','crop_mod2','crop_mod3' ] #,'albmarTRIFFID']#,'crop','crop_mod1']  
Prunids=runids.copy()
#Prunids[-1]='crop_mod2'
PFTindices=[5,5,5,5,5,5,5,5]
#'albmarDATA','std','std_wFIXES','albmarTRIFFID','TRIFFID',
tstep='tstep'
nJULES=len(runids)

BASE_DIR='/prj/GREENHOUSE/GREENHOUSE_sites/'

JULES_OUTPUT_DIR=BASE_DIR+'output/'+site+'/'
JULES_fnames=[JULES_OUTPUT_DIR+JULESvn+'_'+runid+'/'+site+'_'+runid+'.'+tstep+'.nc' \
              for runid in runids]
print(JULES_fnames)

SITE_DIR=BASE_DIR+'data/'+site+'/'
SITE_fname=SITE_DIR+'ConCrop_FullFlux_data.nc'
SITE_fname2=SITE_DIR+'ConCrop_FullFlux_data_albmar.nc'
SITE_laicanht_file=SITE_DIR+'Brattleby_LAI_Canht_data.nc'

```

```python

WK_varnames=['NEE','GPP','TER','SHF','LHF']
WK_soilnames=['T_soil','SMC','SHF_grnd']

SITE_varnames=['NEE','GPP','TER','H','LE']
SITE_soilnames=['T_soil','SMC','downward_heat_flux_at_ground_level']
SITE_timename='time'

JULES_varnames=['npp_gb','gpp_gb','resp_s_gb','resp_p_gb','t_soil','smcl','ftl_gb','latent_heat','fsmc_gb','lai','canht','tstar_gb']
JULES_timename='time'

```

```python
SSinf=nc.Dataset(SITE_fname,'r')
SITE_soil_date=nc.num2date(SSinf.variables[SITE_timename][:], \
                          units=SSinf.variables[SITE_timename].units)
SITE_soil_Dict={}
for wkvar,invar in zip(WK_soilnames,SITE_soilnames):
    SITE_soil_Dict[wkvar]=SSinf.variables[invar][:]
SSinf.close()

SITE_soil_Dict['SMC']*=100.
SITE_soil_Dict['T_soil']  = np.mean(SITE_soil_Dict['T_soil'].reshape(-1,4),axis=1)
SITE_soil_Dict['SMC0']    = np.mean(SITE_soil_Dict['SMC'][...,0],axis=1)
SITE_soil_Dict['SMC1']    = np.mean(SITE_soil_Dict['SMC'][...,1],axis=1)
SITE_soil_Dict['SMC']     = np.mean(SITE_soil_Dict['SMC'].reshape(-1,4),axis=1)
SITE_soil_Dict['SHF_grnd']= np.mean(SITE_soil_Dict['SHF_grnd'],axis=1)

SITE_soil_DF=pd.DataFrame(SITE_soil_Dict,index=SITE_soil_date)
```

```python
SVinf=nc.Dataset(SITE_laicanht_file,'r')
#SITE_veg_date=nc.num2date(np.append(SVinf.variables['date'][:],SVinf.variables['date'][:]+730), \
#                          units=SVinf.variables['date'].units)
SITE_veg_date=nc.num2date(SVinf.variables['date'][:], \
                          units=SVinf.variables['date'].units)
SITE_veg_Dict={}
for var in ['lai','lai_sd','canht']:
    #SITE_veg_Dict[var]=np.ma.masked_equal(np.append(SVinf.variables[var][:],SVinf.variables[var][:]).data,
    #                                      SVinf.variables[var]._FillValue)
    SITE_veg_Dict[var]=SVinf.variables[var][:]

SVinf.close()
SITE_veg_DF=pd.DataFrame(SITE_veg_Dict,index=SITE_veg_date)
```

```python
# Read in Site data and convert to pandas
Sinf=nc.Dataset(SITE_fname,'r')

SITE_Dict={}
for inname,wkname in zip(SITE_varnames,WK_varnames):
    #print(inname,wkname)
    SITE_Dict[wkname]=Sinf.variables[inname][:].squeeze()

SITE_time=nc.num2date(Sinf.variables[SITE_timename][:],  \
                      Sinf.variables[SITE_timename].units)
Sinf.close()
SITE_Dict['NEE']*=-1.0

SITE_panda = pd.DataFrame(SITE_Dict,index=SITE_time)

SITE_time_index = (SITE_panda.index>'2014-09-01')&(SITE_panda.index<'2016-01-01')
SITE_panda = SITE_panda[SITE_time_index]

SITE_panda = SITE_panda.dropna()
```

```python
# Read in Alberto Site data and convert to pandas
Sinf=nc.Dataset(SITE_fname2,'r')

SITE_Dict={}

for wkname in ['NEE','GPP','TER','SHF','LHF']:    #print(inname,wkname)
    SITE_Dict[wkname]=Sinf.variables[wkname][:].squeeze()

SITE_time=nc.num2date(Sinf.variables['date'][:],  \
                      Sinf.variables['date'].units)
Sinf.close()
SITE_Dict['NEE']*=-1.0

SITE_panda2 = pd.DataFrame(SITE_Dict,index=SITE_time)
```

```python
# Read in JULES data and convert to pandas
J_pandas=[]

for Jcnt in range(nJULES):
    #print(runids[Jcnt])
    #print(JULES_fnames[Jcnt])
    iPFT=PFTindices[Jcnt]
    Jinf=nc.Dataset(JULES_fnames[Jcnt],'r')
    
    J_Dict={}
    for inname in JULES_varnames:
        #print(inname)
        J_Dict[inname]=Jinf.variables[inname][:].squeeze()
    
    J_time=nc.num2date(Jinf.variables[SITE_timename][:],  \
                       Jinf.variables[SITE_timename].units)
    Jinf.close()
    
    # Copy some variables straight over
    invars=['npp_gb','gpp_gb','resp_s_gb','resp_p_gb']
    wkvars=['NPP','GPP','RESP_S','RESP_P','SHF','LHF','fsmc']
    JULES_Dict={wkvar:J_Dict[invar]*kgC_to_umolsCO2_factor \
                for wkvar,invar in zip(wkvars,invars)}
    
    invars=['ftl_gb','latent_heat','fsmc_gb','tstar_gb']
    wkvars=['SHF','LHF','fsmc','tstar']
    for invar,wkvar in zip(invars,wkvars):
        JULES_Dict[wkvar]=J_Dict[invar] 
    
    # NEE=NPP-RESP_S
    JULES_Dict['NEE']=JULES_Dict['NPP']-JULES_Dict['RESP_S']
    # TER = RESP_S + RESP_P
    JULES_Dict['TER']=JULES_Dict['RESP_S']+JULES_Dict['RESP_P']
    
    # Top soil layer (top 10cm)
    JULES_Dict['SMC'] = J_Dict['smcl'][:,0]
    JULES_Dict['T_soil'] = J_Dict['t_soil'][:,0]
    for iSoil in range(4):
        JULES_Dict['SMC'+str(iSoil)] = J_Dict['smcl'][:,iSoil]
        JULES_Dict['T_soil'+str(iSoil)] = J_Dict['t_soil'][:,iSoil]
    
    
    # Get specific PFT value for lai and canht
    for var in ['lai','canht']:
        JULES_Dict[var] = J_Dict[var][:,iPFT]
    
    J_pandas.append( pd.DataFrame(JULES_Dict,index=J_time) )
```
