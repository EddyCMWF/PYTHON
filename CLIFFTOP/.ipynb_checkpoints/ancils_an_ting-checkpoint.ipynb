```python
import numpy as np
import netCDF4 as nc

import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(20,12)
%matplotlib inline

```

```python
frac_file='frac.nc'
soil_file='soils_clay.nc'
```

```python
frac=nc.Dataset(frac_file,'r').variables['frac'][:]
frac=np.ma.masked_less(frac,0)
```

```python
soil_inf=nc.Dataset(soil_file,'r')
SOIL_DICT={}
for var in soil_inf.variables:
    SOIL_DICT[var]=soil_inf.variables[var][:]
```

```python
plt.figure(figsize=(15,10))
plt.imshow(frac[-1,:],origin='bottom',interpolation='nearest')
plt.colorbar()
plt.show()
plt.figure(figsize=(15,10))
plt.imshow(np.sum(frac[:-1,:],axis=0),origin='bottom',interpolation='nearest')
plt.colorbar()
plt.show()
frac.shape
```

```python
FIG,AXES = plt.subplots(ncols=3,nrows=3,figsize=(15,6))
for itile in range(9):
    AX=AXES.flatten()[itile]
    AX.imshow(frac[itile,:][40:,:50],origin='bottom',interpolation='nearest',vmin=0,vmax=1)
    AX.set_title(itile)
```
