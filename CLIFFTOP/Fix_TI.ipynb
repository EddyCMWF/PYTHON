```python
import numpy as np
import netCDF4 as nc

import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(20,12)
%matplotlib inline

```

```python
gridfile='grid_info.nc'
grinf=nc.Dataset(gridfile,'r')
lats_2d=grinf.variables['latitude'][:]
lons_2d=grinf.variables['longitude'][:]
grindex=grinf.variables['land_index'][:]
grindex=grindex.astype(int)
grinf.close()
```

```python
infile='timean-tisd-fexp-marthews-imogen.nc'
ammf=nc.Dataset(infile,'a')
```

```python

DATA_DICT={}

for var in ammf.variables:
    DATA_DICT[var]=ammf.variables[var][:]
    
```

```python
var='ti_mean'
plot_data=DATA_DICT[var]

plt.imshow(plot_data,origin='bottom')
plt.colorbar()
plt.show()
```

```python
for var in ['fexp','ti_mean','ti_sd']:
    badex=np.where((grindex.mask==False)&(DATA_DICT[var]<0.))
    goodex=np.where((grindex.mask==False)&(DATA_DICT[var]>=0.))
    n_badpoints=len(badex[0])

    for ibdpt in range(n_badpoints):
        bd_pt = badex[0][ibdpt],badex[1][ibdpt]
        #print(bd_pt)
        
        in_dist_pt=np.argmin( (goodex[0]-bd_pt[0])**2 + (goodex[1]-bd_pt[1])**2 )
        
        #print(min_dist_pt)
        gd_pt=goodex[0][min_dist_pt],goodex[1][min_dist_pt]
        
        DATA_DICT[var][bd_pt]=DATA_DICT[var][gd_pt]
    
    ammf.variables[var][:]=DATA_DICT[var]
    
```

```python
var='fexp'
plot_data=np.ma.masked_less(DATA_DICT[var],0)

plt.imshow(plot_data,origin='bottom')
plt.colorbar()
plt.show()
```
