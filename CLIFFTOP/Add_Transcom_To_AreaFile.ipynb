```python
import netCDF4 as nc
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from scipy import stats

plt.rcParams['figure.figsize']=(20,12)

%matplotlib inline
```

```python
ancil_dir='/prj/CLIFFTOP/COMMON_DATA/ANCILS/'

area_file_1d=ancil_dir+'Area_in_iris_format.nc'
area_file_2d=ancil_dir+'grid_info.nc'

transcom_dir='/prj/wetlands_africa/UM_Modelling/'
transcom_file=transcom_dir+'TRANSCOM_Regions_0.25.nc'

```

```python
#Read Transcom data:
trans_inf = nc.Dataset(transcom_file,'r')
trans_lat = trans_inf.variables['latitude'][:]
trans_lon = trans_inf.variables['longitude'][:]
transcom = trans_inf.variables['transcom_regions'][:]
trans_inf.close()

trans_lon[trans_lon>180] -= 360

trans_lon_2d,trans_lat_2d=np.meshgrid(trans_lon,trans_lat)
```

```python
transcom[transcom==23]=-1
transcom = np.ma.masked_greater(transcom,11)
plt.figure(figsize=[15,11])
plt.imshow(transcom,origin='bottom')
plt.colorbar()
```

```python
#read in the 1d imogen data:
inf=nc.Dataset(area_file_1d,'a')
im_lat=inf.variables['latitude'][:].squeeze()
im_lon=inf.variables['longitude'][:].squeeze()


```

```python
print(trans_lon.shape,trans_lat.shape)
print(im_lat)
print(im_lon)
```

```python
lat_res=2.5
lon_res=3.75
n_impoints=len(im_lat)
fill_value=-999.
im_transcom_1d = np.zeros_like(im_lat)

for ipt in range(n_impoints):
    lat,lon = im_lat[ipt],im_lon[ipt]
    cell_mask = (trans_lat_2d>=lat)&(trans_lat_2d<lat+lat_res)&(trans_lon_2d>=lon)&(trans_lon_2d<lon+lon_res)&(transcom<=11)
    print(ipt,lat,lon,stats.mode(transcom[cell_mask]))
    if len(transcom[cell_mask])>=1:
        im_transcom_1d[ipt] = stats.mode(transcom[cell_mask]).mode[0]
    else:
        im_transcom_1d[ipt] = fill_value 
    
    
badex=np.where(im_transcom_1d==fill_value)[0]
```

```python

print(badex)
print(im_lat[badex])
print(im_lon[badex])

# fix gap fills manually:
fill_1=[6]
im_transcom_1d[badex[fill_1]]=1
fill_2=[4]
im_transcom_1d[badex[fill_2]]=2
fill_3=[1,2,3,5]
im_transcom_1d[badex[fill_3]]=3
fill_4=[0,]
im_transcom_1d[badex[fill_4]]=4
fill_7=[7,]
im_transcom_1d[badex[fill_7]]=7
fill_11=[8,]
im_transcom_1d[badex[fill_11]]=11
fill_m1=[9,]
im_transcom_1d[badex[fill_m1]]=-1
```

```python
# set arctic to positive integer, 12
im_transcom_1d[im_transcom_1d==-1]=12
```

```python
#add transcomm to Area file
outvar=inf.createVariable('transcom','int',('time','y','x'))
outvar.long_name='TRANSCOMM index'
outvar[:] = im_transcom_1d
```

```python
inf.close()
```

```python
# Read in 2d grid info to check and store in 2d
inf_2d=nc.Dataset(area_file_2d,'a')
lat_2d = inf_2d.variables['latitude'][:]
lon_2d = inf_2d.variables['longitude'][:]
land_index = inf_2d.variables['land_index'][:]
```

```python
im_transcom_2d=np.ma.masked_array(im_transcom_1d[land_index],mask=land_index.mask,fill_value=fill_value)
im_transcom_2d[im_transcom_2d.mask==True]=fill_value
plt.figure(figsize=[20,15])
plt.imshow(im_transcom_2d.data,origin='bottom')#,vmin=-10)
plt.colorbar()
```

```python
#add transcomm to Area file
outvar=inf_2d.createVariable('transcom','float32',('y','x'),fill_value=-999)
outvar.long_name='TRANSCOMM index'
outvar[:] = im_transcom_2d
```

```python
inf_2d.close()
```
