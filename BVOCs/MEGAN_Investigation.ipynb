```python
MEGAN_DATA_DIR='/users/eow/edwcom/BVOCs/MEGAN_DATA/'
sub_dir='NA_PFTs/'
species='pft01net08' #'isop22'
```

```python
#species=glob.glob(MEGAN_DATA_DIR+sub_dir+'*')
#print(species)
specie=['pft01net08','pft02ndt08','pft05bet08','pft07bdt08','pft10ds08',\
        'pft12agr08','pft13cgr08','pft14wgr08','pft15ocr08','pft16crn08']
```
