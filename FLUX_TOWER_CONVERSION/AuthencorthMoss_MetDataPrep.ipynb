```python
import netCDF4 as nc
import numpy as np

import matplotlib.pylab as plt


from matplotlib.pyplot import rcParams
%matplotlib inline
rcParams['figure.figsize'] = (15.,10)

from maths_tools import MetTools as MT
```

```python
data_dir='/prj/wetlands_africa/jules/JASMIN/BVOCs/'

infile=data_dir+'AuchencorthMoss_Met_2015_JULES_drive_data.nc'

outfile=data_dir+'AuchencorthMoss_Met_2015_JULES_drive_data_gapfilled.nc'
```

```python
inf=nc.Dataset(infile,'r')
DataDict={str(var): { 'data': inf.variables[var][:].squeeze(), \
                      'units': inf.variables[var].units} \
              for var in ['t','precip','pstar','wind','RH','sw_down','rad_net']}
time=inf.variables['time'][:]
time_obj=nc.num2date( inf.variables['time'][:],units=inf.variables['time'].units )
#inf.close()
```

```python
for var in DataDict:
    print(var, DataDict[var]['data'].shape)
```

```python
Out_DataDict={}
Out_DataDict['time']={'data':time,'units':inf.variables['time'].units}
```

```python
# fill pstar and t with interpolated values real value
for var in ['t','pstar']:
    #fill_value=DataDict[var]['fill_value']
    temp_data=DataDict[var]['data'].copy()
    badex=np.where(temp_data.mask==True)[0]
    goodex=np.where(temp_data.mask==False)[0]
    filled_data=np.interp(time,time[goodex],temp_data[goodex])
    Out_DataDict[var]={'data':filled_data,'units':DataDict[var]['units']}
```

```python
# fill RH with mean diurnal cycle values
for var in ['RH']:
    temp_data=DataDict[var]['data'].reshape(365,48)
    diurnal_cycle = temp_data.mean(axis=0)#

    for ihh in range(48):
        temp_data_hh=temp_data[:,ihh]
        temp_data_hh[temp_data_hh.mask==True]=diurnal_cycle[ihh]
        temp_data[:,ihh]=temp_data_hh
    #
    # Convert to Specific Humidity
    spec_hum=MT.rh2sh(temp_data.flatten(),Out_DataDict['pstar']['data'],Out_DataDict['t']['data'])
    Out_DataDict['q']={'data':spec_hum,'units':'kg kg^-1'}
    
    #print(diurnal_cycle.shape)
    plt.plot(time_obj,spec_hum,label='raw')
    plt.plot(time_obj,MT.rh2sh(DataDict[var]['data'],\
                                 Out_DataDict['pstar']['data'],\
                                 Out_DataDict['t']['data']),label='raw')
    plt.legend()
    plt.show()
    
    #Out_DataDict[var]=temp_RH.flatten()
```

```python
# fill radiation with previous day value
for var in ['rad_net','sw_down']:
    temp_data=DataDict[var]['data'].copy()
    badex=np.where(temp_data.mask==True)[0]
    while len(badex)>0:
        temp_data[badex]=temp_data[badex-48]
        badex=np.where(temp_data.mask==True)[0]
    
    Out_DataDict[var]={'data':temp_data.flatten().copy(),'units':DataDict[var]['units']}
    
    plt.plot(time_obj,temp_data,label='filled')
    plt.plot(time_obj,DataDict[var]['data'],label='raw')
    plt.legend()
    plt.show()
        
```

```python
# Fill precip with zeros
var='precip'

temp_data=DataDict[var]['data'].copy()
temp_data[temp_data.mask==True]=0

plt.plot(time_obj,temp_data,label='filled')
plt.plot(time_obj,DataDict[var]['data'],label='raw')
plt.legend()

Out_DataDict[var]={'data':temp_data,'units':'kg m^-2 s^-1'}
```

```python
# Fill wind with a constant value of 3 m s^-1
var='wind'
Out_DataDict[var]={'data':np.zeros_like(DataDict[var]['data'].data)+3.,'units':'m s^-1'}

```

```python
for var in Out_DataDict:
    print(var)

print(inf.dimensions)
```

```python
#Output data to netCDF
#inf=nc.Dataset(infile,'r')
outf=nc.Dataset(outfile,'w')

outf.createDimension('time',len(inf.dimensions['time']))
outf.createDimension('land',len(inf.dimensions['x']))

for var in Out_DataDict:
    print(var)
    if var=='time':
        outvar=outf.createVariable(str(var),'float32',('time'))
    else:
        outvar=outf.createVariable(str(var),'float32',('time','land'))            
    outvar.units=Out_DataDict[var]['units']
    outvar[:]=Out_DataDict[var]['data']

outf.close()

```
